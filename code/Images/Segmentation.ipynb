```python
import torch
import matplotlib.pyplot as plt
from segment_anything import SamAutomaticMaskGenerator, sam_model_registry
from PIL import Image
import numpy as np

# Load the input image
image_path = "test.jpg"
image = Image.open(image_path)

# Convert the image to a format suitable for SAM
input_image = np.array(image)

# Define model type (use the appropriate type, e.g., 'vit_h')
model_type = "vit_h"  # This must match the architecture used for your checkpoint
sam = sam_model_registry[model_type]()

# Load the custom SAM2 Hiera Large model checkpoint directly
sam_checkpoint = "sam2_hiera_large.pt"
state_dict = torch.load(sam_checkpoint)

# Load the state dictionary into the model
sam.load_state_dict(state_dict)

# Move the model to the appropriate device (CPU or CUDA)
sam.to(device="cuda" if torch.cuda.is_available() else "cpu")

# Create an automatic mask generator
mask_generator = SamAutomaticMaskGenerator(sam)

# Generate the segmentation masks
masks = mask_generator.generate(input_image)

# Visualize the first segmentation mask (you can modify this to combine masks if needed)
segmentation_map = masks[0]['segmentation']

# Plot the input image on the left and segmentation map on the right
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Display input image
axes[0].imshow(image)
axes[0].set_title("Input Image")
axes[0].axis('off')

# Display segmentation map
axes[1].imshow(segmentation_map, cmap="gray")
axes[1].set_title("Segmentation Map")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
