```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
from collections import Counter

# Load a pretrained YOLOv8 model
model = YOLO('yolov8n.pt')

# Load an image
image_path = 'test.jpg'
img = cv2.imread(image_path)

# Perform object detection
results = model(img)

# Get the names of the detected objects and their count
class_names = results[0].names  # Get the class names
detected_classes = results[0].boxes.cls  # Get the class IDs of detected objects

# Count the number of occurrences for each class
class_counts = Counter(detected_classes.tolist())

# Visualize results
annotated_img = results[0].plot()  # Draw the bounding boxes on the image

# Convert BGR (OpenCV format) to RGB (matplotlib format)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Original image conversion
annotated_img_rgb = cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB)  # Annotated image conversion

# Create a figure with two subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Display the original image on the left
axs[0].imshow(img_rgb)
axs[0].axis('off')  # Hide axes
axs[0].set_title('Original Image')

# Display the annotated image on the right
axs[1].imshow(annotated_img_rgb)
axs[1].axis('off')  # Hide axes
axs[1].set_title('YOLOv8 Object Detection')

# Show both images side by side
plt.tight_layout()
plt.show()

# Print out each object type with the count
for class_id, count in class_counts.items():
    object_name = class_names[int(class_id)]  # Get the object name from the class ID
    print(f"Detected {count} {object_name}{'s' if count > 1 else ''}")
```
